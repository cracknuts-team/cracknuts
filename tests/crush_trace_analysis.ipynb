```python
import numpy as np
import zarr
import os

import plotly.graph_objects as go
from plotly_resampler import FigureResampler

# 打开 Zarr 数据集
origin_trace_path = r'D:\project\cracknuts\tests\dataset\42-plain-180.zarr'
zds = zarr.open(origin_trace_path, mode="r")
traces = zds["/0/0/traces"][:180]

# 将所有波形转为 numpy 数组
traces = np.array(traces)

# 计算平均曲线
mean_trace = np.mean(traces, axis=0)

# 创建支持重采样的 Plotly 图形对象
fig = FigureResampler(go.Figure())

# 添加所有灰色曲线
for i, trace in enumerate(traces):
    fig.add_trace(
        go.Scattergl(
            y=trace,
            mode='lines',
            # line=dict(width=1, color='lightgray'),
            line=dict(width=1),
            name=f'Trace {i}'
        )
    )

# 添加红色平均曲线
fig.add_trace(
    go.Scattergl(
        y=mean_trace,
        mode='lines',
        line=dict(width=2, color='green'),
        name='Mean Trace'
    )
)

# 更新布局样式
fig.update_layout(
    height=800,
    plot_bgcolor='white',   # 背景设为白色
    paper_bgcolor='white',
    xaxis=dict(showgrid=True, gridcolor='lightgray'),
    yaxis=dict(showgrid=True, gridcolor='lightgray'),
    title='Traces with Mean (plotly-resampler)'
)

# 显示交互式图形
fig
```
