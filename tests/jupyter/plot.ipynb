```python
import numpy as np
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from cracknuts.trace import ZarrTraceDataset

trace_dataset = ZarrTraceDataset.load(r'D:\work\00.project\cracknuts-show\dataset\stm32f103_aes_em.zarr')
traces = trace_dataset.get_origin_data()['/0/0/traces']
traces = traces[:, :2000]
num_lines = 100
fig = go.Figure()
for i in range(num_lines):
    fig.add_trace(go.Scattergl(y=traces[i], mode='lines', name=f'Line {i+1}', line=dict(width=1)))
fig.update_layout(
    title=f"Plotly 显示 {num_lines} 条曲线",
    showlegend=False,
    xaxis_title="X",
    yaxis_title="Y",
    height=600
)
fig.show()
```
