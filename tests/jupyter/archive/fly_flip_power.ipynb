```python
import numpy as np
import matplotlib.pyplot as plt

def plot_correlation_peaks_plotly_to_matplotlib(correlation_data, bytes_index, the_key, sample_len=None):
    """
    将原来的 Plotly-Resampler 图转换为 Matplotlib 静态绘图版本。
    correlation_data: ndarray, shape = (256, N_bytes, N_samples)
    bytes_index: int, 指定第几个字节
    the_key: int, 目标 key 的索引
    sample_len: int or None, 指定绘制的采样长度
    """

    if sample_len is None:
        sample_len = correlation_data.shape[2]

    x = np.arange(sample_len)

    # 创建画布
    fig, ax = plt.subplots(figsize=(9, 3))  # 约等于 900x300 像素
    fig.patch.set_facecolor("white")

    # 绘制所有 key 曲线（非 target key 灰色）
    for i in range(256):
        if i == the_key:
            continue
        ax.plot(
            x,
            correlation_data[i, bytes_index, :sample_len],
            color="gray",
            linewidth=0.5,
            alpha=0.3
        )

    # 绘制 target key（红色）
    ax.plot(
        x,
        correlation_data[the_key, bytes_index, :sample_len],
        color="red",
        linewidth=1.0,
        label=f"target key {the_key}"
    )

    # 样式与布局
    ax.set_facecolor("white")
    ax.grid(True, linewidth=1, color="#dddddd", linestyle="-", alpha=0.6)
    ax.set_xlabel("Sample index")
    ax.set_ylabel("Correlation")
    ax.set_title(f"Correlation peaks for byte {bytes_index} (key {the_key})", pad=10)
    ax.legend(loc="upper right", frameon=False)

    # 边框样式（类似 Plotly 的 mirror=True）
    for spine in ax.spines.values():
        spine.set_linewidth(1)
        spine.set_color("#dddddd")

    plt.tight_layout()
    plt.show()

    return fig, ax
```

```python
plot_correlation_peaks_plotly_to_matplotlib(zd["/0/0/correlation"], 0, 0x11)
```

```python
import plotly.graph_objects as go
from plotly_resampler import FigureWidgetResampler
import zarr

zd = zarr.open(r'D:\work\00.project\cracknuts-show\dataset\smt32f103_aes_power_CPAAnalysis.zarr')
data = zd["/0/0/correlation"][:,0,:]

fig = FigureWidgetResampler(go.Figure())

# ======== 3. 添加每条曲线 ========
for i in range(data.shape[0]):
    if i == 0x11:
        continue
    fig.add_trace(
        go.Scattergl(
            y=data[i, :],
            mode='lines',
            name=f"Signal {i+1}",
            line=dict(
                color='gray',   # 可以是颜色名 'red', 'blue', 'green' 或者十六进制 '#ff0000'
                width=1        # 线宽，可选
            )
        )
    )

fig.add_trace(
        go.Scattergl(
            y=data[0x11, :],
            mode='lines',
            name=f"Signal {i+1}",
            line=dict(
                color='red',   # 可以是颜色名 'red', 'blue', 'green' 或者十六进制 '#ff0000'
                width=1        # 线宽，可选
            )
        )
    )

fig.update_layout(
    title="二维 NumPy 数据示例",
    width=2000,         # ← 指定画布宽度（像素）
    height=500
)

fig
```
