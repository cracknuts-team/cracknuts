```python
import cracknuts as cn
```

```python
def bytes_bin(b: bytes, sep: str = ' ') -> str:
    return sep.join(f'{byte:08b}' for byte in b)
```

```python
s1 = cn.new_cracker('192.168.0.29')
s1.connect(force_update_bin=True, force_write_default_config=True)
```

```python
s1.get_firmware_version()
```

```python
# 复位 PL IO
s1.register_write(base_address=0x43c10000, offset=0x300, data=0x01)
```

```python
# 使能 PL IO
s1.register_write(base_address=0x43c10000, offset=0x304, data=0b0000_0000_0000_0000_0000_0000_0000_0000)
s1.register_write(base_address=0x43c10000, offset=0x304, data=0b0000_0000_0000_0000_0000_0000_0001_1111)
```

```python
# 输入输出控制
s1.register_write(base_address=0x43c10000, offset=0x308, data=0b0000_0000_0000_0000_0000_0000_0000_0000) # 输出
# s1.register_write(base_address=0x43c10000, offset=0x308, data=0b0000_0000_0000_0000_0000_0000_0000_1111) # 输入
```

```python
s1.register_write(base_address=0x43c10000, offset=0x318, data=0b0000_0000_0000_0000_0000_0000_0000_1111) # io0 四段高电平
s1.register_write(base_address=0x43c10000, offset=0x31C, data=0b0000_0000_0000_0000_0000_0000_0000_1111) # io1 四段高电平
s1.register_write(base_address=0x43c10000, offset=0x320, data=0b0000_0000_0000_0000_0000_0000_0000_1111) # io2 四段高电平
s1.register_write(base_address=0x43c10000, offset=0x324, data=0b0000_0000_0000_0000_0000_0000_0000_1111) # io3 四段高电平
```

```python
# io0设置波形长度
s1.register_write(base_address=0x43c10000, offset=0x32c, data=250_000_000) 
s1.register_write(base_address=0x43c10000, offset=0x330, data=250_000_000) 
s1.register_write(base_address=0x43c10000, offset=0x334, data=250_000_000) 
s1.register_write(base_address=0x43c10000, offset=0x338, data=250_000_000) 
# io1设置波形长度
s1.register_write(base_address=0x43c10000, offset=0x33C, data=250_000_000) 
s1.register_write(base_address=0x43c10000, offset=0x340, data=250_000_000) 
s1.register_write(base_address=0x43c10000, offset=0x344, data=250_000_000) 
s1.register_write(base_address=0x43c10000, offset=0x348, data=250_000_000) 
# io2设置波形长度
s1.register_write(base_address=0x43c10000, offset=0x34c, data=250_000_000) 
s1.register_write(base_address=0x43c10000, offset=0x350, data=250_000_000) 
s1.register_write(base_address=0x43c10000, offset=0x354, data=250_000_000) 
s1.register_write(base_address=0x43c10000, offset=0x358, data=250_000_000) 
# io3设置波形长度
s1.register_write(base_address=0x43c10000, offset=0x35c, data=250_000_000) 
s1.register_write(base_address=0x43c10000, offset=0x360, data=250_000_000) 
s1.register_write(base_address=0x43c10000, offset=0x364, data=250_000_000) 
s1.register_write(base_address=0x43c10000, offset=0x368, data=250_000_000) 
```

```python
# 设置波形开始
s1.register_write(base_address=0x43c10000, offset=0x30c, data=0b0000_0000_0000_0000_0000_0000_0001_1111)
```

```python
import time
while True:
    s1.register_write(base_address=0x43c10000, offset=0x30c, data=0b0000_0000_0000_0000_0000_0000_0001_1111)
    # 10 秒 通电，等待5秒
    time.sleep(15)
```

```python
import random
import time
from cracknuts.cracker import serial

def init(c):
    # 复位 PL IO
    s1.register_write(base_address=0x43c10000, offset=0x300, data=0x01)
    # 使能 PL IO
    s1.register_write(base_address=0x43c10000, offset=0x304, data=0b0000_0000_0000_0000_0000_0000_0000_0000)
    s1.register_write(base_address=0x43c10000, offset=0x304, data=0b0000_0000_0000_0000_0000_0000_0001_1111)
    # 输入输出控制
    s1.register_write(base_address=0x43c10000, offset=0x308, data=0b0000_0000_0000_0000_0000_0000_0001_0000) # 0-3 输出，5输入
    # 设置四个io四段高电平
    s1.register_write(base_address=0x43c10000, offset=0x318, data=0b0000_0000_0000_0000_0000_0000_0000_1111) # io0 四段高电平
    s1.register_write(base_address=0x43c10000, offset=0x31C, data=0b0000_0000_0000_0000_0000_0000_0000_1111) # io1 四段高电平
    s1.register_write(base_address=0x43c10000, offset=0x320, data=0b0000_0000_0000_0000_0000_0000_0000_1111) # io2 四段高电平
    s1.register_write(base_address=0x43c10000, offset=0x324, data=0b0000_0000_0000_0000_0000_0000_0000_1111) # io3 四段高电平
    # 设置四个io高电平总时长为10s
    # io0设置波形长度
    s1.register_write(base_address=0x43c10000, offset=0x32c, data=250_000_000) 
    s1.register_write(base_address=0x43c10000, offset=0x330, data=250_000_000) 
    s1.register_write(base_address=0x43c10000, offset=0x334, data=250_000_000) 
    s1.register_write(base_address=0x43c10000, offset=0x338, data=250_000_000) 
    # io1设置波形长度
    s1.register_write(base_address=0x43c10000, offset=0x33C, data=250_000_000) 
    s1.register_write(base_address=0x43c10000, offset=0x340, data=250_000_000) 
    s1.register_write(base_address=0x43c10000, offset=0x344, data=250_000_000) 
    s1.register_write(base_address=0x43c10000, offset=0x348, data=250_000_000) 
    # io2设置波形长度
    s1.register_write(base_address=0x43c10000, offset=0x34c, data=250_000_000) 
    s1.register_write(base_address=0x43c10000, offset=0x350, data=250_000_000) 
    s1.register_write(base_address=0x43c10000, offset=0x354, data=250_000_000) 
    s1.register_write(base_address=0x43c10000, offset=0x358, data=250_000_000) 
    # io3设置波形长度
    s1.register_write(base_address=0x43c10000, offset=0x35c, data=250_000_000) 
    s1.register_write(base_address=0x43c10000, offset=0x360, data=250_000_000) 
    s1.register_write(base_address=0x43c10000, offset=0x364, data=250_000_000) 
    s1.register_write(base_address=0x43c10000, offset=0x368, data=250_000_000) 

   # 设置输入使能
    s1.register_write(base_address=0x43c10000, offset=0x310, data=0b0000_0000_0000_0000_0000_0000_0001_0000) # mask使能
    s1.register_write(base_address=0x43c10000, offset=0x314, data=0b0000_0000_0000_0000_0000_0000_0000_0001) # 或 模式

    s1.osc_sample_clock('8m')
    s1.osc_sample_length(80_000_000)
    s1.osc_sample_delay(0)
    s1.osc_analog_enable('A')
    s1.osc_analog_disable('B')
    s1.osc_analog_gain('A', 20)

def do(cracer, count):
    s1.register_write(base_address=0x43c10000, offset=0x30c, data=0b0000_0000_0000_0000_0000_0000_0001_1111)
    # 10 秒 通电，等待5秒
    time.sleep(20)
    
    return {
        "plaintext": b'',
        "ciphertext": b'',
        "key": b''
    }


def finish(c):
    ...
    # print('optional behavior')


acq = cn.new_acquisition(s1, init=init, do=do, finish=finish)
cn.panel(acq)
```
