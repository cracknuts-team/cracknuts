```python
## 时钟glitch配置
import math
# 配置normal时钟
# "64 M"
# "24 M"
# "12 M"
# "8  M"
# "4  M"
clock_len = 10
g1.register_write(base_address=0x43c10000, offset=0x1D0, data=clock_len) # g1.glitch_clock_normal()
for i in range(clock_len):
    if i < clock_len/2:
        voltage_code = 0x0000
    else:
        voltage_code = 0x0294 # 3.3v
    g1.register_write(base_address=0x43c10000, offset=0x1D4, data=voltage_code) # g1.glitch_clock_wave_narmal()

# 一个时间减半，幅度减半的时钟
glitch_clock_len = 20
g1.register_write(base_address=0x43c10000, offset=0x1DC, data=glitch_clock_len) # g1.glitch_clock_len()
for i in range(glitch_clock_len):
    if i < glitch_clock_len/2:
        voltage_code = 0x0000
    else:
        voltage_code = 0x0134  # 1.5v
    g1.register_write(base_address=0x43c10000, offset=0x1E0, data=voltage_code) # g1.glitch_clock_wave()
g1.register_write(base_address=0x43c10000, offset=0x1D8, data=1)  # glitch 等待    
g1.register_write(base_address=0x43c10000, offset=0x1E4, data=2)  # glitch 间隔
g1.register_write(base_address=0x43c10000, offset=0x1E8, data=2)  # glitch 重复次数

g1.register_write(base_address=0x43c10000, offset=0X1C8, data=1) #
```
