```python
import cracknuts as cn
```

```python
s1 = cn.cracker_s1('192.168.0.15')
```

```python
s1.connect(force_update_bin=True, force_write_default_config=True)
```

```python
s, r = s1.register_read(base_address=0x43C10000, offset=0x0C04)
r.hex()
```

```python
s1.register_write(base_address=0x43C10000, offset=0x0C04, data=0b10) # 1 读取 0 输出
```

```python
s1.register_write(base_address=0x43C10000, offset=0x0C04, data=0x00) # 1 读取 0 输出
```

```python
s1.register_write(base_address=0x43C10000, offset=0x0C00, data=0b00)
```

```python
s, r = s1.register_read(base_address=0x43C10000, offset=0x0C00)
r
```

```python
' '.join(format(byte, '08b') for byte in r)
```

```python
s1.digital_read(5),s1.digital_read(6)
```

```python
s1.digital_pin_mode(5, 1)
s1.digital_pin_mode(6, 1)
```

```python
s1.digital_pin_mode(5, 'INPUT')
s1.digital_pin_mode(6, 'INPUT')
```
