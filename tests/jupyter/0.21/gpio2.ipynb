```python
def get_bit(data: bytes, bit_index: int) -> int:
    """
    从 bytes 对象 data 中读取第 bit_index 位（从 0 开始，整体按字节顺序，
    每个字节内部：bit 0 是最低位 LSB）
    
    例如：data = b'\x05' (即 0b00000101)
    - bit_index=0 → 1
    - bit_index=2 → 1
    - bit_index=1 → 0
    """
    if bit_index < 0 or bit_index >= len(data) * 8:
        raise IndexError("bit_index out of range")
    
    byte_index = bit_index // 8
    bit_in_byte = bit_index % 8
    
    byte_value = data[byte_index]
    return (byte_value >> bit_in_byte) & 1
```

```python
def get_bit_lsb(data: bytes, bit_index: int) -> int:
    byte_index = bit_index // 8
    if byte_index >= len(data):
        raise IndexError

    bit_in_byte = bit_index % 8
    return (data[byte_index] >> bit_in_byte) & 1

```

```python
def get_bit_stream_lsb(data: bytes, bit_index: int) -> int:
    value = int.from_bytes(data, byteorder="big")
    return (value >> bit_index) & 1
```
